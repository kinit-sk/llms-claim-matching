```python
import pandas as pd
import regex as re
from glob import glob
```

```python
paths = glob('./results/*.csv')
for path in paths:
    if 'xlt' in path or 'cot' in path:
        df = pd.read_csv(path)
        df['prediction_label'] = ''
        
        for index, row in df.iterrows():
            prediction = row['Prediction']
            if re.search(':.{0,15}Yes', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Oui', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Tak', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search('Answer: Not Relevant', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('Answer: Partially Relevant', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search("answer is \\'Yes\\'", prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':(.{0,5})\nYes', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}No', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}Nie', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('：{0,15}No', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('not relevant', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('irrelevant', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('relevant', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Nein', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}Ναι', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Όχι', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}Не', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}Да', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Няма', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}Igen', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Ja', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}لا', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}हाँ', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}नहीं।', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}नहीं', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}نعم', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Ano', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif prediction.startswith('Да'):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search('nem releváns', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('संबंधित नहीं है', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('संबंध नहीं रखता है', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('nie jest relewantne', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('और दोनों में इसके निर्माण और चालु होने की बात कही गई है', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search('संबंधित नाही', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('संबंधित', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search('साथ पूरी तरह से प्रासंगिक', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Nee', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('is related to', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}नो', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('सीधे तौर पर जुड़ा हुआ नहीं है।', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search('प्रासंगिक नहीं', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('does not directly address the claim', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('not related to', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('There is no overlap between', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search('There is no direct connection', prediction):
                df.at[index, 'prediction_label'] = 'No'
            elif re.search(':.{0,15}是', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            elif re.search(':.{0,15}Partial Relevance', prediction):
                df.at[index, 'prediction_label'] = 'Yes'
            else:
                df.at[index, 'prediction_label'] = 'No'
                
        df.to_csv(path, index=False)
```
